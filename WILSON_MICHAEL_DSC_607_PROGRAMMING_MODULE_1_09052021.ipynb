```python
'''The Obesity Data dataset is formally titled "CDC Data: Nutrition, Physical Activity, & Obesity" and is a dataset 
generated from the Behavioral Risk Factor Surveillance System of percentages of the population estimated to be 
engaged in certain activities, behaviors, or classifications relating to weight, physical activity, and dietary 
habits based on survey results gathered by the US Centers for Disease Control (CDC).  The data reported are annual 
statistics for the weight classification questions, and bi-annual for the other question variables.  These data 
points were reported by state/territory and nationwide, from 2011 to 2016.
gathered along with the numerical data are categorical breakouts of the question responses both overall, 
and by gender, age bracket, income bracket, race/ethnicity and edcational attainment level.  This dataset was found 
on Kaggle, posted by community member spittman1248, at 
https://www.kaggle.com/spittman1248/cdc-data-nutrition-physical-activity-obesity (Pittman, 2018).

Once the data was downloaded and imported, summary statistics were generated using the describe() method in pandas.
'''
```

```python
#load packages

import pandas as pd
import numpy as np

ObesityData = pd.read_csv(r'C:\Users\Mike\Documents\Grad School 2021\DSC-607 Data Mining\Module 1\Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv') #read in the dataset

ObesityData.describe(include = 'all')  #Generate summary statistics on dataset, including categorical columns.
```

```python
ObesityData.describe() #Summary statistics for numeric variables
```

```python
'''Data Preparation:  For the model, we are not interested in the categorical breakdowns, other than state total
values, which appear in the Data_Value column'''

#Drop unnecessary columns
ObesityData2 = ObesityData.drop(['YearStart','LocationDesc','Class','Data_Value_Unit','Data_Value_Type',
                                 'Data_Value_Alt','Data_Value_Footnote_Symbol','Data_Value_Footnote','GeoLocation',
                                 'Low_Confidence_Limit','High_Confidence_Limit ','Age(years)','Education',
                                'Gender','Income','Race/Ethnicity','DataValueTypeID'], axis=1)

ObesityData2.describe()
```

```python
#Drop unnecessary rows, which are any row for which the data value reported is not a state-wide total number.

ObesityData3 = ObesityData2.dropna(subset=['Total']) #Drop any data not classified as a total

ObesityData3 = ObesityData3[ObesityData3['LocationAbbr'] != 'US']  #Eliminate nation-level data from set


ObesityData3.describe()
```

```python
'''Take a look at box plots of response values by question'''

%matplotlib inline
Response_Range_Boxplot = ObesityData3.boxplot(column = ['Data_Value'], by = ['Question'], figsize = (40,20), 
                                                rot = 90, fontsize = 20)
```

```python
'''Check to see if any states or territories have provided a different amount of data than others'''

ObesityData3.value_counts(['LocationAbbr']).plot(kind = 'bar', figsize = (25, 10),
                                                   title = 'Number of responses per state or territory',
                                                  xlabel = 'State/Territory',
                                                  ylabel = 'Number of responses')
```

```python
'''Histogram to check spread of sample sizes used for reporting the data values, sorted by question'''

ObesityData3.hist(column = 'Sample_Size', bins = 20, by = 'Question', layout = (10,1), figsize = (25,100))
```

```python
'''Need to pivot data so that the questions are columns of the response data values.'''
ObesityData4 = ObesityData3.pivot(index =['LocationAbbr','YearEnd'], columns = 'Question', values = 'Data_Value')

ObesityData4.head()  #Check to make sure the data is arranged as needed
```

```python
'''All of the questions other than the two identifying a weight status, are only reported on odd years
Let's discard those rows, as there's no data with which to predict the target value'''

ObesityData5 = ObesityData4.dropna() #Drop any rows with NaN values
ObesityData5.describe() #Checking to see data counts are the same for all columns
```

```python
'''For the general linear model, I am interested in trying to predict the overweight classification rate 
statistic using the data response values for the questions of eating vegetables less than once daily, and the 
percent reported to not engage in any lesiure-time physical activity'''

#Load statsmodels packages

import statsmodels.api as sm

X = ObesityData5[['Percent of adults who engage in no leisure-time physical activity',
              'Percent of adults who report consuming vegetables less than one time daily']] #predictors

Y = ObesityData5['Percent of adults aged 18 years and older who have an overweight classification'] #target

X = sm.add_constant(X, prepend=False) #add a constant value (intercept)
 
Obesitymodel = sm.OLS(Y, X).fit()
predictions = Obesitymodel.predict(X) 
 
Obesitymodel.summary()




```

```python
'''Conclusion:  The model report indicates that the dependent variable of Percent of adults aged 18 years and
older who have an overweight classification can be predicted by the independent variables of Percent of adults 
who engage in no leisure-time physical activity and Percent of adults who report consuming vegetables less than 
one time daily, but not very well.  While the model is significant, the variable relating to no leisure-time 
physical activities would not be considered significant at a p <=0.01 rejection threshold.  More importantly,
a very low adjusted R-squared statistic of 0.071 means the model is only able to explain about 7% of
the variation in the target.'''

'''Reference:
Pittman, S. (2018) "CDC Data: Nutrition, Physical Activity, & Obesity - Obesity Trends in US "(Version 1)
    [Data set]. https://www.kaggle.com/spittman1248/cdc-data-nutrition-physical-activity-obesity'''
```
